## 키워드에 대한 검색량 데이터 수집 함수화

```python
def get_header(method, uri, api_key, secret_key, customer_id):
    timestamp = str(round(time.time() * 1000))
    signature = signaturehelper.Signature.generate(timestamp, method, uri, SECRET_KEY)
    
    return {'Content-Type': 'application/json; charset=UTF-8', 
            'X-Timestamp': timestamp, 'X-API-KEY': API_KEY, 
            'X-Customer': str(CUSTOMER_ID), 'X-Signature': signature}

def get_keyword_info(keyword):
    uri = '/keywordstool'
    method = 'GET'
    r = requests.get(BASE_URL + uri + f'?hintKeywords={keyword}&showDetail=1',
                     headers=get_header(method, uri, API_KEY, SECRET_KEY, CUSTOMER_ID))
    return pd.DataFrame(r.json()['keywordList'])
```

## 검색어에 대한 상품등록수 수집 함수화

```python
def item_count(keyword):
    url = f'https://search.shopping.naver.com/search/all?query={keyword}'
    req = requests.get(url)
    response = TextResponse(req.url, body=req.text, encoding="utf-8")
    count = response.xpath('//*[@id="__next"]/div/div[2]/div/div[3]/div[1]/div[1]/ul/li[1]/a/span[1]/text()').extract()
    
    if count is None:
        count = response.xpath('//*[@id="__next"]/div/div[2]/div/div[3]/div[1]/div[1]/ul/li[1]/a/span[1]/text()').extract()
        
    return ''.join(count)
```

```python
keyword_list = keywordList[['relKeyword', 'monthlyPcQcCnt', 'monthlyMobileQcCnt']].copy()
keyword_list = keyword_list[:20]
```

```python
datas = []

for relKeyword in keyword_list['relKeyword']:
    datas.append(item_count(relKeyword))
```

```python
item_cnt = pd.DataFrame(datas, columns=['item_count'])
```

```python
item_count = pd.concat([keyword_list,item_cnt],axis=1)

item_count.head()
```

```python
item_count['total_search_count'] = item_count['monthlyPcQcCnt'] + item_count['monthlyMobileQcCnt']
```

```python
item_count['item_count'] = item_count['item_count'].apply(lambda x: x.replace(',', ''))
```
